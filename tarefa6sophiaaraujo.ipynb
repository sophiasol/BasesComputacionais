Tensão em função do tempo.

```python
import numpy as np
import matplotlib.pyplot as plt
 
t = np.arange (0, 10, 0.05)

v = np.e ** (-0.5 * t) * np.cos(2 * np.pi * 0.8 * t)

print (v)


plt.figure()
plt.plot(t,v, color='pink')
plt.show()
```

Equivalência da temperatura em Fahrenheits em função da temperatura em Celsius:

```python
import numpy as np
import matplotlib.pyplot as plt

c = np.linspace (-20, 100, 100)

f = 9/5 * c + 32

print (f)

plt.figure()
plt.plot(c,f, color='purple')
plt.show()
```

Crescimento populacional por tempo:

```python
import numpy as np
import matplotlib.pyplot as plt

a= np.linspace(1994, 2022, 28)

p = 800000 *np.e ** ((a-1994)/40)

print (p)

plt.figure()
plt.plot(p,a, color='red')
plt.show()
```
